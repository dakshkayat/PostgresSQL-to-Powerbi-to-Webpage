```python
import random
from faker import Faker
import pandas as pd
import datetime as dt

# Set up the Faker instance with Indian locale
faker = Faker('en_IN')

# Number of rows to generate
num_rows = 600

# List to hold the generated customer data
customer_data = []

# Function to generate a phone number in the 10-digit format
def generate_phone_number():
    return ''.join([str(random.randint(7, 9)) for _ in range(1)]) + ''.join([str(random.randint(0, 9)) for _ in range(9)])

# Function to generate a gmail email address
def generate_gmail_email(firstname, lastname):
    return f"{firstname.lower()}.{lastname.lower()}@gmail.com"

# Loop to generate customer data
for i in range(1, num_rows + 1):
    # Generate random first name and last name
    firstname = faker.first_name()
    lastname = faker.last_name()

    # Generate random date of birth between 18 and 70 years old
    dob = faker.date_of_birth(minimum_age=18, maximum_age=50)

    # Generate random city
    city = faker.city()

    # Define country as India
    country = "India"

    # Generate a random postal code (Indian pattern)
    postal_code = faker.postcode()

    # Generate a phone number in the desired format
    phone = generate_phone_number()

    # Generate a Gmail email address
    email = generate_gmail_email(firstname, lastname)
    
    # Calculate the date when the customer turns 18
    dob_plus_18 = dob + dt.timedelta(days=18*365)

    # Generate "DateEntered" to be after they turned 18
    date_entered = faker.date_between_dates(
        date_start=dob_plus_18, date_end=dt.date.today()
    )

    # Append the generated data as a tuple
    customer_data.append(
        (i, firstname, lastname, dob, city, country, postal_code, phone, email, date_entered)
    )

# Create a DataFrame from the generated data
df = pd.DataFrame(
    customer_data,
    columns=[
        "customerid",
        "firstname",
        "lastname",
        "Date_of_Birth",
        "City",
        "Country",
        "postal_code",
        "phone",
        "email",
        "DateEntered",
    ],
)

# Save the generated data to a CSV file
df.to_csv('Customers.csv')
```

```python
import random
from faker import Faker
import pandas as pd
import datetime as dt

# Set up the Faker instance
faker = Faker()

# Number of rows for products
num_products = 300


# Generate unique product data
product_categories = {
    "Electronics": ["Smartphone", "Laptop", "Tablet", "Headphones", "Camera"],
    "Furniture": ["Sofa", "Table", "Chair", "Bed", "Desk"],
    "Clothing": ["T-shirt", "Jeans", "Jacket", "Dress", "Sweater"],
    "Sports": ["Football", "Basketball", "Tennis Racket", "Cricket Bat", "Baseball Glove"],
    "Toys": ["Action Figure", "Doll", "Puzzle", "Lego Set", "Remote Control Car"],
    "Beauty": ["Lipstick", "Foundation", "Eyeliner", "Blush", "Perfume"],
}

# Generate unique product data with category-related names
product_data = []
for i in range(1, num_products + 1):
    productid = i
    product_category = random.choice(list(product_categories.keys()))
    product_name = random.choice(product_categories[product_category])
    price_per_unit = random.uniform(10, 1000)  # A random price per unit between 10 and 1000
    stock_available = random.randint(10, 500)  # A random stock count
    supplier = faker.company()  # A random company name as the supplier

    product_data.append(
        (productid, product_name, product_category, price_per_unit, stock_available, supplier)
    )

# Create a DataFrame for the products table
df_products = pd.DataFrame(
    product_data,
    columns=[
        "productid",
        "product_name",
        "Product_category",
        "price_per_unit",
        "stock_available",
        "supplier",
    ],
)
# Save the products table to a CSV file
df_products.to_csv('Products.csv')

```
